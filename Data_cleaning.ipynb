```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
with open('newdata/json/254.json','r') as f:
    d = json.load(f)
    #print(d)
    Xcar=[]
    ycar=[]
    for i in d['drawings']['fish']:
        #print(len(i))
        #print(i)
        for pt in i:
            #print(pt[0])
            Xcar.append(pt[0])
            ycar.append(pt[1])
    #print(pts)
```

```python
fig, ax = plt.subplots()

# Plot the data
ax.plot(Xcar, ycar)

# Invert the Y-axis to display it right-side up
ax.invert_yaxis()

# Show the plot
plt.show()
```
